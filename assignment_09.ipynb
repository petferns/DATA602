```python
import plotly.express as px
import plotly.io as pio
pio.renderers.default = 'iframe'
```

```python
df = px.data.tips()
df.head()
```

```python
bar1 = px.bar(df, x="sex", y="total_bill", color="smoker", barmode="group")
bar1.show("png")
```

```python
bar2 = px.bar(df, x="sex", y="total_bill", color="day", barmode="group")
bar2.show("png")
```

```python
scatter1 = px.scatter(df, x="total_bill", y="tip", color="sex", facet_col="smoker")
scatter1.show("png")
```

```python
scatter2 = px.scatter(df, x="total_bill", y="tip", color="sex", facet_col="smoker", trendline="ols")
scatter2.show("png")
```

```python
scatter3 = px.scatter(df, x="total_bill", y="tip", facet_col="day", facet_row="time", 
                category_orders={'day': ['Thur', 'Fri', 'Sat', 'Sun'], 'time': ['Dinner', 'Lunch']})
scatter3.show("png")
```

```python
hist = px.histogram(df, x="tip", marginal="rug")
hist.show("png")
```

```python
box1 = px.box(df, y="tip",x="smoker", color="smoker")
box1.show("png")
```
